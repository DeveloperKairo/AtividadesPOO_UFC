Algoritmo verifica se os pontos inseridos pelo usuario formam um triangulo ou um quadrilatero e os classifica

```python
import math

def distancia(p1, p2):
    return int(math.sqrt((p2[0] - p1[0])**2 + (p2[1] - p1[1])**2))

def classificar_triangulo(lados):
    segmento1, segmento2, segmento3 = lados
    if math.isclose(segmento1, segmento2) and math.isclose(segmento2, segmento3):
        return 'Equilátero'
    elif math.isclose(segmento1, segmento2) or math.isclose(segmento1, segmento3) or math.isclose(segmento2, segmento3):
        return 'Isóceles'
    else:
        return 'Escaleno'

def classificar_quadrilatero(lados):
    segmento1, segmento2, segmento3, segmento4 = lados
    if (math.isclose(segmento1, segmento2) and
        math.isclose(segmento2, segmento3) and
        math.isclose(segmento3, segmento4)):
        return 'Quadrado'
    elif math.isclose(segmento1, segmento3) and math.isclose(segmento2, segmento4):
        return 'Retângulo'
    else:
        return 'Não identificado/Não é um quadrilatero'

pontos = []
classificar = int(input("Digite 3 para triângulo ou 4 para quadrilátero: "))

if classificar == 3:
    for i in range(3):
        x = float(input(f"Digite a coordenada x do {i + 1}º ponto: "))
        y = float(input(f"Digite a coordenada y do {i + 1}º ponto: "))
        pontos.append((x, y))

    lados = [distancia(pontos[i], pontos[(i+1)%3]) for i in range(3)]

    if (lados[0] < lados[1] + lados[2] and
        lados[1] < lados[0] + lados[2] and
        lados[2] < lados[0] + lados[1]):
        classe = classificar_triangulo(lados)
        print(f'O triângulo é {classe}.')
    else:
        print('Não é um triângulo!')

elif classificar == 4:
    for i in range(4):
        x = float(input(f"Digite a coordenada x do {i + 1}º ponto: "))
        y = float(input(f"Digite a coordenada y do {i + 1}º ponto: "))
        pontos.append((x, y))

    lados = [distancia(pontos[i], pontos[(i+1)%4]) for i in range(4)]
    classe = classificar_quadrilatero(lados)
    print(f'O quadrilátero é {classe}.')
else:
    print('Valor inválido!')
```

Algoritmo gera pontos para formar um triangulo equilatero, escaleno ou isosceles

```python
import random
import math
import matplotlib.pyplot as plt

def pontos_equilatero():
    x0, y0 = random.uniform(0, 10), random.uniform(0, 10)
    L = random.uniform(1, 5)
    A = (x0, y0)
    B = (x0 + L, y0)
    C = (x0 + L/2, y0 + L * math.sqrt(3)/2)
    return A, B, C

def pontos_isosceles():
    x0, y0 = random.uniform(0, 10), random.uniform(0, 10)
    base = random.uniform(2, 6)
    altura = random.uniform(1, 5)
    A = (x0, y0)
    B = (x0 + base, y0)
    C = (x0 + base/2, y0 + altura)
    return A, B, C

def pontos_escaleno():
    x0, y0 = random.uniform(0, 10), random.uniform(0, 10)
    dx1, dy1 = random.uniform(2, 6), random.uniform(1, 5)
    dx2, dy2 = random.uniform(3, 7), random.uniform(2, 6)
    A = (x0, y0)
    B = (x0 + dx1, y0 + dy1)
    C = (x0 + dx2, y0 + dy2)
    return A, B, C

print('Escolha a classificação do triângulo:')
print('1 - Equilátero')
print('2 - Isósceles')
print('3 - Escaleno')
opcao = int(input('Digite o número correspondente: '))

if opcao == 1:
    pontos = pontos_equilatero()
    print('Pontos do triângulo equilátero:')
elif opcao == 2:
    pontos = pontos_isosceles()
    print('Pontos do triângulo isósceles:')
elif opcao == 3:
    pontos = pontos_escaleno()
    print('Pontos do triângulo escaleno:')
else:
    print('Opção inválida!')
    pontos = None

if pontos:
    print(f'A = {pontos[0]}')
    print(f'B = {pontos[1]}')
    print(f'C = {pontos[2]}')

x_coords = [pontos[0][0], pontos[1][0], pontos[2][0], pontos[0][0]]  # Fecha o triângulo
y_coords = [pontos[0][1], pontos[1][1], pontos[2][1], pontos[0][1]]

plt.figure(figsize=(6,6))
plt.plot(x_coords, y_coords, 'b-', label='Lados do triângulo')
plt.scatter([p[0] for p in pontos], [p[1] for p in pontos], color='red', s=80, zorder=5)
plt.text(pontos[0][0], pontos[0][1], 'A', fontsize=12, ha='right')
plt.text(pontos[1][0], pontos[1][1], 'B', fontsize=12, ha='left')
plt.text(pontos[2][0], pontos[2][1], 'C', fontsize=12, ha='left')
plt.title('Triângulo formado pelos pontos')
plt.xlabel('Eixo X')
plt.ylabel('Eixo Y')
plt.grid(True)
plt.legend()
plt.axis('equal')
plt.show()
```
